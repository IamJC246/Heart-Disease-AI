```python
#converts alphabetical texts to numbers where,

# Sex
# M = 0
# F = 1 

# Chest pain type
# ATA = 1
# NAP = 2
# ASY = 3
# TA = 4

# Resting ECG
# Normal = 0 
# ST = 1
# LVH = 2

# Exercise Angina
# N = 0
# Y = 1

#ST Slope
# Up = 1
# Flat = 0
# Down = 2

dfCopy['Sex'] = dfCopy['Sex'].replace(['M'],'0')
dfCopy['Sex'] = dfCopy['Sex'].replace(['F'],'1')

dfCopy['ChestPainType'] = dfCopy['ChestPainType'].replace(['ATA'],'1')
dfCopy['ChestPainType'] = dfCopy['ChestPainType'].replace(['NAP'],'2')
dfCopy['ChestPainType'] = dfCopy['ChestPainType'].replace(['ASY'],'3')
dfCopy['ChestPainType'] = dfCopy['ChestPainType'].replace(['TA'],'4')

dfCopy['RestingECG'] = dfCopy['RestingECG'].replace(['Normal'],'0')
dfCopy['RestingECG'] = dfCopy['RestingECG'].replace(['ST'],'1')
dfCopy['RestingECG'] = dfCopy['RestingECG'].replace(['LVH'],'2')

dfCopy['ExerciseAngina'] = dfCopy['ExerciseAngina'].replace(['N'],'0')
dfCopy['ExerciseAngina'] = dfCopy['ExerciseAngina'].replace(['Y'],'1')

dfCopy['ST_Slope'] = dfCopy['ST_Slope'].replace(['Flat'],'0')
dfCopy['ST_Slope'] = dfCopy['ST_Slope'].replace(['Up'],'1')
dfCopy['ST_Slope'] = dfCopy['ST_Slope'].replace(['Down'],'2')
```

```python
# to determine the outliers
dfCopy_outliers = pd.DataFrame(dfCopy.loc[:,])
```

```python
# Calculate IQR
Q1 = dfCopy_outliers.quantile(0.25)
Q3 = dfCopy_outliers.quantile(0.75)
IQR = Q3 - Q1

print(IQR)
```

```python
# to use IQR score to filter out the outliers by keeping only valid values

# to replace every outlier on the upper side by the upper whisker 
for i, j in zip(np.where(dfCopy_outliers > Q3 + 1.5 * IQR)[0], np.where(dfCopy_outliers > Q3 + 1.5 * IQR)[1]):
    
    whisker  = Q3 + 1.5 * IQR
    dfCopy_outliers.iloc[i,j] = whisker[j]
    
# to replace every outlier on the lower side by the lower whisker 
for i, j in zip(np.where(dfCopy_outliers < Q1 - 1.5 * IQR)[0], np.where(dfCopy_outliers < Q1 - 1.5 * IQR)[1]): 
    
    whisker  = Q1 - 1.5 * IQR
    dfCopy_outliers.iloc[i,j] = whisker[j]
```

```python
# to remove all columns
dfCopy.drop(columns =  dfCopy.loc[:,], inplace = True)
```

```python
# to concatenate the new columns in the dataset where outliers were replaced with whisker values
dfCopy = pd.concat([dfCopy , dfCopy_outliers], axis = 1)
```

```python
# to show the boxplot of the dataset where outliers were replaced with whisker values
dfCopy.boxplot(column = cols, rot=45, figsize = (20,10))
```

### 4. Categorical Data Encoding

```python
dfCopy_clean
```

```python
# To create the matrix of independent variable, x
X = dfCopy_clean.iloc[:,0:10].values
# To create the matrix of dependent variable, y
Y = dfCopy_clean.iloc[:,10:11].values
```

```python
X
```

```python
Y
```

```python
# To encode the categorical data
from sklearn.preprocessing import LabelEncoder

# A. To encode the categorical data (State) in the independent variable, X
label_encoder_X = LabelEncoder()
X[:,0] = label_encoder_X.fit_transform(X[:,0])
```

```python
# To split the whole dataset into training dataset and testing dataset
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0) #train_size=0.8, you can either still put this or not since test_size is already defined. By default, remaining is for training
```

### 5. Feature standardization

```python
from sklearn.preprocessing import StandardScaler # for not normally distributed samples
standard_scaler = StandardScaler ()
```

### 6. Feature Selection

```python
X_train_standard = X_train.copy()
X_test_standard = X_test.copy()
X_train_standard = standard_scaler.fit_transform(X_train_standard)
X_test_standard = standard_scaler.fit_transform(X_test_standard) 
```

### Logistic Regression with validations

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(solver ='lbfgs', 
                           max_iter = 100,
                           multi_class = 'auto',
                           verbose = 0,
                           warm_start = False,
                           random_state=0, 
                           n_jobs = -1) 
```

```python
model.fit(X_train_standard,Y_train.ravel())
```

```python
Y_predict = model.predict(X_test_standard)
Y_predict
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(Y_test, Y_predict)

import seaborn as sns
sns.heatmap(confusion_matrix, annot=True)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Value')
plt.ylabel('Actual Value')

TP = confusion_matrix[1,1]
TN = confusion_matrix[0,0]
FP = confusion_matrix[0,1]
FN = confusion_matrix[1,0]
```

```python
from sklearn.model_selection import KFold
k_fold = KFold(n_splits=10, shuffle=True, random_state=0) # shuffle = True para sa small dataset

from sklearn.model_selection import cross_val_score

# To feature scale the X_dummytrap Using Standardization
X_standard = X.copy()
X_standard = standard_scaler.fit_transform(X_standard)
```

```python
# For the accuracy as scoring for for cross-validation 
accuracies = cross_val_score(estimator=model, X=X_standard, y=Y.ravel(), cv=k_fold, scoring='accuracy')

accuracies_average = accuracies.mean()
accuracies_variance = accuracies.std()

print('Accuracy of K-FOLDS:')
print (accuracies)
print(' ')
print('Average Accuracy of K-FOLDS:')
print(accuracies_average)
print(' ')
print('Accuracy Variance of K-FOLDS:')
print(accuracies_variance)
print(' ')
```

```python
# To evaluate the performance of the logistic regression model using holdout
# A. For the Classification Accuracy
from sklearn.metrics import accuracy_score
classification_accuracy = accuracy_score(Y_test, Y_predict)
print('Classification Accuracy: %.4f'
      % classification_accuracy)
print(' ')

# B. For the Classification Error
from sklearn.metrics import accuracy_score
classification_error = 1-classification_accuracy
print('Classification Error: %.4f'
      % classification_error)
print(' ')

# C. For the Sensitivity or Recall Score / True Positive Rate (If looking at the positive, how often does positive occur? Performance of predicting + and - should be the same to consider the model without bias towards one class
# True Positive Rate:  Actual Value +, how often Correct

Recall = TP/(TP+FN)
print('Recall/Sensitivity: %.4f'
      % Recall)
print(' ')

# D. For the Specficity (If looking at the negative actual, how often does negative occur? Performance of predicting + and - should be the same to consider the model without bias towards one class. 
# Counterpart of sensitivity)
# True Negative Rate: Actual Value -, how often correct
specificity = TN/(TN+FP)
print('Specificity: %.4f'
      % specificity)
print(' ') 

# E. For the FP rate .
# False Positive Rate: Actual Value -, how often inorrect
false_positve_rate = 1-specificity
print('False Positive Rate: %.4f'
      % false_positve_rate)
print(' ')  

# F. For the precision.

precision = TP/(TP+FP)
print('Precision: %.4f'
      % precision)
print(' ')  

# G. For the F1 score. Relating precision and sensitivity
# False Negative Rate: Predicted Value and Actual Value+, how often the prediction is correct relation


F1_Score = 2*(Recall*precision)/(Recall+precision)
print('F1 Score: %.4f'
      % F1_Score)
print(' ')  
```
